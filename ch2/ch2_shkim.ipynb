## 2. 퍼셉트론

#### 1957년 프랭크 로젠블라트가 고안한 알고리즘으로, 딥러닝의 기원이 되는 아이디어. <br>
* 퍼셉트론 : 다수의 신호를 입력받아 하나의 신호 출력
* 퍼셉트론 신호는 0(신호 x) 또는 1(신호 o) 값을 가진다.

)

* 입력 신호가 뉴런에 보내질 때 가중치(w)가 곱해짐. 입력신호의 총합이 임계값(theta)를 넘어갈 때에만 1을 출력.

* 입력 신호 각각에 가중치가 부여됨.
* 가중치가 클 수록 해당 입력 신호가 중요함.

#### 2.2. AND/NAND/OR 게이트(논리 회로)
* 퍼셉트론으로 상기 논리 회로를 표현할 수 있음.
* 모든 경우에 있어서 퍼셉트론의 구조는 동일함.
* 단, 가중치(w)와 임계값(theta)만 적절히 조절하면 됨.





### 2.3 퍼셉트론 구현하기

```
def AND(x1, x2):
  w1, w2, theta = 0.5, 0.5, 0.7 #매개변수 초기화
  tmp = x1*w1 + x2*w2           #tmp = 가중치를 곱한 입력의 총합
  if tmp <= theta:
    return 0
  elif tmp > theta:
    return 1
  
```

```
print(AND(0,0))
print(AND(1,0))
print(AND(0,1))
print(AND(1,1))
```

```
import numpy as np

x = np.array([0,1]) #원소로 0,1이 들어 있는 1차원 배열, 입력값
w = np.array([0.5,0.5]) # 각 신호에 곱해질 가중치
b = -0.7 #편향 (theta)

# (입력 신호 * 가중치 + 편향) 의 값이 0보다 크면 == 1, 그렇지 않으면 == 0

print(w*x) #입력 신호 * 가중치
np.sum(w*x) #입력 신호 * 가중치의 총합
np.sum(w*x)+b #입력 신호 * 가중치의 총합에 편향을 더한 값
```

### 2.3.3 가중치와 편향을 도입하여 AND 게이트 구현하기


```
def AND(x1,x2):
  x = np.array([x1,x2])   #매개변수를 넘파이 배열로 
  w = np.array([0.5,0.5]) #가중치
  b = -0.7                #편향
  tmp = np.sum(w*x) + b  
  if tmp <= 0 :
    return 0
  else:
    return 1
```

* 가중치와 편향은 다른 기능을 한다
* 가중치 : x1, x2.. 각 입력 신호가 결과에 주는 영향력, 즉 중요도를 결정
* 편향 : 뉴런의 활성화 감도(결과 = 1)를 결정
* 예) 편향이 -0.1일 경우
* 입력신호*가중치의 총합이 0.1을 넘을 때에만 뉴런이 활성화됨
* 예) 편향이 -20.0일 경우
* 입력신호*가중치의 총합이 20.0을 넘을 때에만 뉴런이 활성화됨

NAND/OR 게이트

```
def NAND(x1,x2):
  x = np.array([x1,x2])
  w = np.array([-0.5,-0.5])  #AND와는 가중치와 편향만 다르다 
  b = 0.7
  tmp = np.sum(w*x)+b
  if tmp <=0:
    return 0
  else:
    return 1

def OR(x1,x2):
  x = np.array([x1,x2])
  w = np.array([0.5,0.5])
  b = -0.2
  tmp = np.sum(w*x)+b
  if tmp <=0:
    return 0
  else:
    return 1
```

```
print("--NAND--")
print(NAND(0,0))
print(NAND(1,0))
print(NAND(0,1))
print(NAND(1,1))

print("--OR---")

print(OR(0,0))
print(OR(1,0))
print(OR(0,1))
print(OR(1,1))
```

### 퍼셉트론의 한계 - XOR 게이트
* 단층 퍼셉트론은 직선 하나로 나눈 영역만 표현할 수 있다. 
* 이것을 해결한 것이 '다층 퍼셉트론 multi-layer perceptron'이다.
* 퍼셉트론의 층을 쌓으면서 비선형적인 영역을 분리할 수 있게 되었다.




```
#XOR 게이트 구현하기

def XOR(x1,x2):
  s1 = NAND(x1,x2)
  s2 = OR(x1,x2)
  y = AND(s1,s2)
  return y
```

```
print(XOR(0,0))
print(XOR(1,0))
print(XOR(0,1))
print(XOR(1,1))
```
